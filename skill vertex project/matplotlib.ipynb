```python
import matplotlib.pyplot as plt

```

```python
x = [10,20,30,40]
y = [20,30,40,50]
plt.plot(x,y)
plt.title("Example")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
x = [3,1,3]
y = [3,2,1]
plt.plot(x,y)
plt.title("Line Chart")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
a = [3,1,3,12,2,4,4]
b = [3,2,1,4,5,6,7]
plt.bar(a,b)
plt.title("Bar chart")
plt.legend(['items'])
plt.show()
```

```python
x = [1,2,3,4,5,6,7,3]
plt.hist(x, bins=[1,2,3,4,5,6,7])
plt.title("Histogram")
plt.show()
```

```python
a = [3,1,3,12,2,4,4]
b = [3,2,1,4,5,6,7]
plt.scatter(a,b)
plt.title("Scatter plot")
plt.show()
```

```python
import numpy as np
x = np.array([35,25,25,15])
cols = ["black", "yellow","blue","red"]
plt.pie(x, explode=(0,0.1,0,0), labels=["Maths", "English", "Science", "Social"], startangle=90, colors = cols)
plt.title("Pie Chart")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [1,8,27,64,125]
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot3D(z ,y ,x)
```

```python
import seaborn as sns

```

```python
sns.set_theme()
```

```python

sns.distplot([0, 1, 2, 3, 4, 5])

plt.show()
```

```python
sns.distplot([0, 1, 2, 3, 4, 5], hist=False)

plt.show()
```

```python
from numpy import random
sns.distplot(random.normal(loc=50, scale=7, size=1000), hist=False, label='normal')
sns.distplot(random.poisson(lam=50, size=1000), hist=False, label='poisson')

plt.show()
```

```python
sns.distplot(random.uniform(size=1000), hist=False)

plt.show()
```
