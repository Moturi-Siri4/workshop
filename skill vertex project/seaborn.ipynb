```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("miniconda3\data\Population.csv")
df
```

```python
sns.lineplot(x="Year", y ="Population", data=df)
plt.title("title")
sns.set_style("dark")
plt.show()
```

```python
pal = sns.color_palette()
sns.palplot(pal)
plt.show()
```

```python
pal2 = sns.dark_palette("xkcd:golden", 8)
sns.palplot(pal2)
plt.show()
```

```python
sns.set_palette(pal2)
plt.subplot(211)
sns.lineplot(x="Year", y="Population", data=df)
sns.set_palette(pal)
plt.subplot(212)
sns.lineplot(x="Year", y="Population", data=df)
plt.show()
```

```python
total_sales = [3000, 2245, 1235, 5330, 4200]
explode = [0, 0, 0, 0.2, 0]

location = ['Bangalore', 'Delhi', 'Chennai', 'Mumbai', 'Kolkatta']
colors = sns.husl_palette(h=.5)
plt.pie(total_sales, labels=location, colors=colors,
        autopct='%.0f%%', explode=explode, shadow=True, rotatelabels='true')
plt.show()
```

```python
sns.barplot(x="Year", y="Population", data=df)
sns.set_palette(sns.cubehelix_palette(hue=1))
plt.show()
```

```python
sns.histplot(x="Year", y="Population", data=df)
plt.show()
```

```python
sns.distplot(df["Population"])
sns.set_palette(sns.husl_palette(l=.4))
plt.show()
```

```python
sns.jointplot(x="Year", y="Population", data=df)
plt.show()
```

```python
sns.pairplot(data=df, hue="Year")
plt.show()
```
