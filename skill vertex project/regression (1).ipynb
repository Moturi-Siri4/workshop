```python
plt.figure(figsize=(25, 6))

plt.subplot(1,2,1)
plt1 = df.Population.value_counts().plot(kind='bar')
plt.title('Population Histogram')
plt1.set(xlabel = 'Population', ylabel='Frequency of Population')

plt.subplot(1,2,2)
plt1 = df.Year.value_counts().plot(kind='bar')
plt.title('Year Histogram')
plt1.set(xlabel = 'Year', ylabel='Frequency of Year')
```

```python
def scatter(x,fig):
    plt.subplot(5,2,fig)
    plt.scatter(df[x],df['Population'])
    plt.title(x+' vs Population')
    plt.ylabel('Population')
    plt.xlabel(x)

plt.figure(figsize=(10,20))

scatter('Year', 1)
scatter('Median Age', 2)
scatter('Fertility Rate',3)
scatter('World Population',4)


plt.tight_layout()
```
